```python
## H6: Create a simple calculator that performs basic arithmetic operations (addition, subtraction, multiplication, division) based on user input.
```

```python
number1=float(input("Enter first number1:"))
number2=float(input("Enter first number2:"))
operator=(input("Enter an operator +,-,*,/ :"))

if operator=="+":
  result= number1+number2
  print("Addition of the numbers is {}".format(result) )
elif operator=="-":
  result=number1-number2
  print("Substraction of the numbers is {}".format(result) )
elif operator=="*":
  result=number1*number2
  print("Multiplication of the numbers is {}".format(result) )
elif operator=="/":
  result=number1/number2
  print("Division of the numbers is {}".format(result) )
else:
  print("Select correct operator " )
```

```python
## H8: Create a program that converts temperatures between Fahrenheit and Celsius based on user input.
```

```python
temp=float(input("Please enter temperature value :  "))
unit= input("Please enter unit name(C/F) :  ")

if unit.upper()=="C":
  Fahrenheit=(temp * 9/5) + 32
  print("Mentioned Celsius equal to :{:.2f} degree Fahrenheit".format(Fahrenheit))
elif unit.upper()=="F":
  Celsius=(temp - 32) * 5/9
  print("Mentioned Fahrenheit equal to :{:.2f}degree  Celsius".format(Celsius))
else:
  print("not in scope")



```
